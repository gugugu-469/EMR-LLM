```python
log_history = datas['log_history']
```

```python
print(read_file)
x = []
y_1 = []
y_2 = []
y_3 = []
y_1_key = 'reward'

try:
    for log in log_history:
        y_1.append(log[y_1_key])
        x.append(log['step'])
except:
    pass

plt.plot(x, y_1, label=y_1_key)


# 添加标签和标题
plt.xlabel('steps')
plt.ylabel('value')
plt.title('reward curve')
plt.legend()

# 显示图形
plt.show()
```
